```python
import pandas as pd

# Define the data
data = [
    {"name": "ali", "location": "library", "motive": "revenge", "weapon": "knife", "evidence": "none"},
    {"name": "muhammad", "location": "bookstore", "motive": "none", "weapon": "none", "evidence": "none"},
    {"name": "ahmed", "location": "home", "motive": "none", "weapon": "none", "evidence": "none"},
    {"name": "john", "location": "library", "motive": "revenge", "weapon": "gun", "evidence": "bloodstains"},
]

# Convert to DataFrame
df = pd.DataFrame(data)

# Save as CSV
df.to_csv("suspect.csv", index=False)

print("suspect.csv has been created.")
```

```python
# Filtering: Show only those at the library
library_suspects = df[df['location'] == 'library']
print("Suspects at the library:")
display(library_suspects)

# Counting: How many had each weapon type
weapon_counts = df['weapon'].value_counts()
print("\nCounts of weapon types:")
display(weapon_counts)

# Grouping: Suspects by motive
grouped_by_motive = df.groupby('motive')
print("\nSuspects grouped by motive:")
for motive, group in grouped_by_motive:
    print(f"\nMotive: {motive}")
    display(group)
```

#**Part A**

In this cell, the essential libraries are imported that'll be used thorugh out the Colab Notebook. A list is randomly generated and its statistical analysis is displayed. A product of a random 5x5 matrix with its transpose is also displayed.

```python
# Importing Essential Libraries
import numpy as np
import random as rnd
import matplotlib.pyplot as plt

# Part 1
list = [] # Declaring empty list

for i in range(100):
  list.append(rnd.randint(1,100)) # Appends random number between 1 and 100 everytime the loop iterates

# Statistical Measures
print(f"The original list is {list}")
print(f"The average of the list is {np.mean(list)}")
print(f"The median of the list is {np.median(list)}")
print(f"The standard deviation of the list is {np.std(list)}")

list.sort() # Sorts list in ascending order

print(f"The sorted list is {list}")
print(f"The top 5 largest values in the list are {list[-1:-6:-1]}")

print("")

# Part 2
one_d_array = np.arange(25) # 1D array
matrix_5x5 = one_d_array.reshape(5, 5) # Transforming the array into 5x5 matrix
np.fill_diagonal(matrix_5x5, 0) # Replace the diagonal values with 0
transpose = np.transpose(matrix_5x5)
product = np.dot(matrix_5x5, transpose) # Dot product of matrix and its transpose

print(f"The original matrix is {matrix_5x5}\n")
print(f"The product of the matrix with its transpose is {product}\n")

# Part 3
random_values = [] # Empty list

for i in range(50):
  random_values.append(rnd.randint(-5,5))

postive_count = len([i for i in random_values if i > 0]) # Count of positive values
negative_count = len([i for i in random_values if i < 0]) # Count of negative values
num_neg_one_and_pos_one = len([i for i in random_values if i <1 or i >-1]) # Count of -1 and 1


print(f"The random list generated is {random_values}")
print(f"The number of positive values in the list is {postive_count}")
print(f"The number of negative values in the list is {negative_count}")
print(f"The number of values equal to -1 or 1 in the list is {num_neg_one_and_pos_one}")
```

#**Part B**

In this cell, iris dataset is loaded which is a built-in dataset provided by the sklearn library.

```python
# Part 1
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target # Add the target data (species) to the DataFrame

display(df.head(10))
```

In this cell, a new column is added which intends to display the area of a petal.

```python
df['petal_area'] = df['petal length (cm)'] * df['petal width (cm)']
df.head()

filtered_df = df[(df['sepal length (cm)'] > 5) & (df['petal_area'] < 1.5)]
filtered_df['petal length (cm)'].nlargest(5)
```

In this cell, we introduce some null values and later, handle them.

```python
df.loc[df.sample(5).index, 'sepal length (cm)'] = np.nan # Sets 5 random values to Nan
df.isna().sum()
df = df.fillna(df.median(numeric_only=True))
```

#**PART C**

In the following cells, all the graphical analysis (e.g. Charts and Plots) are displayed.

```python
# Histogram
plt.hist(df['petal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], color='blue', alpha=0.7)
plt.title("Scatter Plot: Petal Length vs Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# Calculate mean sepal length per species
# Map the target (species) integers to their names
species_names = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(species_names)

mean_sepal_length = df.groupby('species')['sepal length (cm)'].mean()

# Plot bar chart
plt.bar(mean_sepal_length.index, mean_sepal_length.values, color=['red', 'green', 'blue'])
plt.title("Mean Sepal Length per Species")
plt.xlabel("Species")
plt.ylabel("Mean Sepal Length (cm)")
plt.show()
```

```python
# Boxplot using Pandas + Matplotlib
df.boxplot(column='sepal width (cm)', by='species', grid=False)
plt.title("Sepal Width Distribution per Species")
plt.suptitle("")  # removes automatic title
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

#**Bonus Task**

```python
# Compute correlation matrix
correlation_matrix = df.drop(columns="species").corr()

# Plot heatmap with Matplotlib
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label="Correlation")
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=45, ha="right")
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```
